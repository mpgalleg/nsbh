```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
solar = True
p = np.round(10**np.arange(-0.1,3.5, 0.25),2)

m_donor= 10
step=.2
q_ = np.arange(0.3, 3.5+step, step)
m_bh = np.round(q_*m_donor,2)

print(m_bh)
print(len(m_bh))

grid = {}
m2 = []
p_orb_days = []
for bh in m_bh:
    m2.append(np.ones(len(p))*bh)
    p_orb_days.append(p)
    
grid['m1'] = m_donor
grid['m2'] = np.concatenate(m2)
grid['initial_period_in_days'] = np.concatenate(p_orb_days)
grid['newton_iterations_limit'] = 20
grid['CE_alpha_th'] = 1
grid['CE_alpha'] = 1

if solar:
    grid['new_Z'] = 0.0142
    grid['new_Y'] = 0.2703
    grid['Zbase'] = 0.0142
else:
    grid['new_Z'] = 0.0014
    grid['new_Y'] = 0.2511
    grid['Zbase'] = 0.0014


grid_df = pd.DataFrame(grid)


fig = plt.figure(figsize=(4,6))
plt.scatter(grid_df['m2'].values, grid_df['initial_period_in_days'].values, alpha=0.5)
plt.yscale('log')
```
